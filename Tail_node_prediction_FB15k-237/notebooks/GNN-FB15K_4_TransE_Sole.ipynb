```python
# Pykeen pipeline , transE sole model without GNN
from pykeen.pipeline import pipeline
import torch


pipeline_result_transe = pipeline(
    dataset='FB15k237',
    model='TransE',
    # using L1 norm, as it's a direct metric for TransE
    model_kwargs=dict(
        embedding_dim=512,      # Matched to RotatE
        scoring_fct_norm=1      
    ),
    loss='NSSA',                # Self-Adversarial Loss (Matched to RotatE)
    loss_kwargs=dict(
        margin=9.0,             # Matched to RotatE
        adversarial_temperature=1.0 
    ),
    training_kwargs=dict(
        num_epochs=250,         
        batch_size=1024,
    ),
    optimizer='Adam',
    optimizer_kwargs=dict(lr=0.0005), # Slightly slower LR for 512D space
    negative_sampler='BasicNegativeSampler',
    negative_sampler_kwargs=dict(num_negs_per_pos=128), 
    stopper='early',
    stopper_kwargs=dict(frequency=5, patience=10, relative_delta=0.002),
    device='cuda' if torch.cuda.is_available() else 'cpu',
    random_seed=42,
)
print("TransE Training Complete!")
```

## Metrics(RotatE model and TransE Model)


```python
import os
import torch
import pandas as pd
import numpy as np
from pykeen.datasets import FB15k237
from tqdm.auto import tqdm

dataset = FB15k237()
train_triples = dataset.training.mapped_triples
test_triples = dataset.testing.mapped_triples

num_nodes = dataset.training.num_entities
num_test_edges = test_triples.shape[0]


# Find the exact path for RotatE (Kaggle might nest it slightly depending on the ZIP)
rotate_dir = '/kaggle/input/datasets/ullasgirish/rotate-sota-backup/RotatE_SOTA_Backup-2'
path_rotate = None
for root, dirs, files in os.walk(rotate_dir):
    if 'trained_model.pkl' in files:
        path_rotate = os.path.join(root, 'trained_model.pkl')
        break

if not path_rotate:
    raise FileNotFoundError("Could not find trained_model.pkl for RotatE! Check the Input folder structure.")

# TransE is right where we left it in the working directory
path_transe = './TransE_SOTA_Model/trained_model.pkl'


# Load models (weights_only=False for PyTorch 2.6+ compatibility)
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

model_rotate = torch.load(path_rotate, map_location=device, weights_only=False).eval()
model_transe = torch.load(path_transe, map_location=device, weights_only=False).eval()

# Graph topology and multiplicity
print("\nAnalyzing Graph Structure")
# A. Topology (Node Degrees)
heads, rels, tails = train_triples[:, 0], train_triples[:, 1], train_triples[:, 2]
node_degrees = torch.bincount(heads, minlength=num_nodes) + torch.bincount(tails, minlength=num_nodes)

# B. Multiplicity (1-to-1, 1-to-N, etc.)
df_train = pd.DataFrame({'h': heads.numpy(), 'r': rels.numpy(), 't': tails.numpy()})
tph = df_train.groupby(['r', 'h'])['t'].nunique().groupby('r').mean()
hpt = df_train.groupby(['r', 't'])['h'].nunique().groupby('r').mean()

relation_types = {}
for r in df_train['r'].unique():
    t_ratio, h_ratio = tph.get(r, 1.0), hpt.get(r, 1.0)
    if h_ratio < 1.5 and t_ratio < 1.5: relation_types[r] = '1-to-1'
    elif h_ratio < 1.5 and t_ratio >= 1.5: relation_types[r] = '1-to-N'
    elif h_ratio >= 1.5 and t_ratio < 1.5: relation_types[r] = 'N-to-1'
    else: relation_types[r] = 'N-to-N'


import torch
from tqdm.auto import tqdm

def get_pykeen_ranks(model, test_triples):
    ranks = []
    
    # THE FIX: Dropped from 512 to 32 to prevent 28GB RAM spikes
    batch_size = 32 
    
    # Flush any leftover memory from the crash
    torch.cuda.empty_cache() 

    with torch.no_grad():
        for i in tqdm(range(0, num_test_edges, batch_size), desc=f"Scoring {model.__class__.__name__}"):
            batch = test_triples[i:i+batch_size]
            
            # Send batch to GPU
            hr_batch = batch[:, :2].to(device)
            target_tails = batch[:, 2].to(device)
            
            # Get scores for all 14,541 possible tails
            scores = model.score_t(hr_batch)
            
            # Sort and find the rank of the true tail
            sorted_indices = torch.argsort(scores, dim=1, descending=True)
            for j in range(batch.shape[0]):
                rank = (sorted_indices[j] == target_tails[j]).nonzero(as_tuple=True)[0].item() + 1
                ranks.append(rank)
                
    return ranks

print("\nScoring RotatE...")
ranks_rotate = get_pykeen_ranks(model_rotate, test_triples)

print("\nScoring TransE...")
ranks_transe = get_pykeen_ranks(model_transe, test_triples)

# Bucketing
buckets_rotate = {'Global': [], 'Rare (<=10)': [], 'Hub (>=100)': [], '1-to-1': [], '1-to-N': [], 'N-to-1': [], 'N-to-N': []}
buckets_transe = {'Global': [], 'Rare (<=10)': [], 'Hub (>=100)': [], '1-to-1': [], '1-to-N': [], 'N-to-1': [], 'N-to-N': []}

test_rels = test_triples[:, 1].numpy()
test_tails = test_triples[:, 2].numpy()

for i in range(num_test_edges):
    r = test_rels[i]
    target = test_tails[i]
    degree = node_degrees[target].item()
    rel_type = relation_types.get(r, 'N-to-N')

    r_rot = ranks_rotate[i]
    r_trn = ranks_transe[i]

    # Assign Global
    buckets_rotate['Global'].append(r_rot)
    buckets_transe['Global'].append(r_trn)

    # Assign Topology
    if degree <= 10:
        buckets_rotate['Rare (<=10)'].append(r_rot)
        buckets_transe['Rare (<=10)'].append(r_trn)
    elif degree >= 100:
        buckets_rotate['Hub (>=100)'].append(r_rot)
        buckets_transe['Hub (>=100)'].append(r_trn)

    # Assign Multiplicity
    buckets_rotate[rel_type].append(r_rot)
    buckets_transe[rel_type].append(r_trn)

# Metrics
def calc_metrics(rank_list):
    arr = np.array(rank_list)
    if len(arr) == 0: return 0.0, 0.0, 0.0, 0.0, 0
    mrr = np.mean(1.0 / arr)
    h1 = np.mean(arr <= 1) * 100
    h3 = np.mean(arr <= 3) * 100
    h10 = np.mean(arr <= 10) * 100
    return round(mrr, 4), round(h1, 2), round(h3, 2), round(h10, 2), len(arr)

report_data = []
categories = ['Global', 'Rare (<=10)', 'Hub (>=100)', '1-to-1', '1-to-N', 'N-to-1', 'N-to-N']

for cat in categories:
    mrr_r, h1_r, h3_r, h10_r, count_r = calc_metrics(buckets_rotate[cat])
    report_data.append(['RotatE (SOTA)', cat, count_r, mrr_r, h1_r, h3_r, h10_r])

    mrr_t, h1_t, h3_t, h10_t, count_t = calc_metrics(buckets_transe[cat])
    report_data.append(['TransE (SOTA)', cat, count_t, mrr_t, h1_t, h3_t, h10_t])

df_report = pd.DataFrame(report_data, columns=['Model', 'Category', 'Test_Sample_Size', 'MRR', 'Hits@1(%)', 'Hits@3(%)', 'Hits@10(%)'])
df_report = df_report.sort_values(by=['Category', 'Model'])

print("\n" + "="*85)
print(df_report.to_string(index=False))
print("="*85 + "\n")

file_name = 'pykeen_master_diagnostics.csv'
df_report.to_csv(file_name, index=False)
```
